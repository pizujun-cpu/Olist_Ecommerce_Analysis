```python
#导入库和读取数据
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.pyplot import rcParams
plt.rcParams['font.sans-serif'] = ['SimHei']

# 读取核心文件
df_orders = pd.read_csv('../data/olist_orders_dataset.csv')
df_items = pd.read_csv('../data/olist_order_items_dataset.csv')
df_customers = pd.read_csv('../data/olist_customers_dataset.csv')

# 查看数据大小
print(f"订单表形状：{df_orders.shape}")
print(f"订单详情表形状：{df_items.shape}")
print(f"用户表形状：{df_customers.shape}")
```

### 数据清洗与合并


```python
#1.处理时间格式，把字符串转换成时间
df_orders['order_purchase_timestamp'] = pd.to_datetime(df_orders['order_purchase_timestamp'])
```

```python
#2.合并表（用order_id连接订单和详情，用customer_id连接用户）
#先连订单和详情
df_master = pd.merge(df_orders,df_items,on='order_id',how='inner')
#再连用户信息
df_master = pd.merge(df_master,df_customers,on='customer_id',how='inner')
```

```python
#3.筛选核心字段（减小不必要的内存占用）
core_columns = [
    'customer_unique_id','order_id','order_purchase_timestamp',
    'price','product_id'
]
df = df_master[core_columns].copy()
```

```python
#4.去重和缺失值处理
df.isnull().sum()
```

```python
df = df.dropna(subset=['customer_unique_id'])
```

```python
print(f'清理后数据量:{len(df)}')
```

```python
#将清理后的数据保存到csv中
df.to_csv('../data/clean_data.csv',index=False)
```
